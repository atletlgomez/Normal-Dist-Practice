#Guided Practice 4.3

Use Tom's ACT score, 24, along with the ACT mean and standard deviation to find his Z-score

ACT mean = 21, ACT std = 6

z score = ( x - mu ) / sigma

Tom's Score

```python
tom_zscore = (24 - 21) / 6
tom_zscore
```

Tom's Zscore is 0.5, half a std away from the average but still fall in the first std area

Tom's percentile

```python
norm.cdf(tom_zscore) * 100
```

Tom has score in one standard deviation away from the average

```python
mu = 21
sigma = 6

#creates the line between 3 stds above and below
x = np.linspace(3, 39)

#probablity on the right (y-axis)
y = norm.pdf(x, mu, sigma)


#stops the fill
x_fill = np.linspace(0, mu + sigma * tom_zscore)

#creates fill to the line, uses x_fill to know where to stop
y_fill = norm.pdf(x_fill, mu, sigma)

plt.plot(x, y, label = "normal", color = "green" )
plt.title('Normal Distribution')
plt.xlabel('ACT Score')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.fill_between(x_fill, y_fill, color = "blue", alpha = 0.5)
plt.show()
```

Tom performed better than 69.14% of the class

```python
100 - norm.cdf(tom_zscore) * 100
```

Only 30.85% perfomed better than Tom
